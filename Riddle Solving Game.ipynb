```python
# Import necessary modules

import time  # Module for time-related functions
import threading  # Module for creating and managing threads

# Define a function to get user input with a timeout

def get_input_with_timeout(prompt, timeout):
    
    print(prompt)  # Display the prompt for user input
    input_text = [""]  # Initialize a list to store user input

    # Define a nested function to read input in a separate thread
    def input_thread():
        input_text[0] = input().strip()  # Read user input and strip whitespace

    thread = threading.Thread(target=input_thread)  # Create a thread for input_thread
    thread.start()  # Start the thread to read user input
    thread.join(timeout)  # Wait for the thread to finish or timeout

    return input_text[0]  # Return the user's input

# Define the main function to play the riddle-solving game

def play_game():
    print("Welcome to the Riddle-solving Game!")

    while True:  # Loop to play the game repeatedly
        print("\nScenario 1: You find yourself standing in front of a closed door.")
        print("To open the door, you must solve a riddle.")
        print("\nThe more you take, the more you leave behind. What am I?")
        print("\nA) Footsteps")
        print("b) Memories")
        print("c) Darkness")

        while True:  # Loop for user input with timeout
            
            # Get user input using the get_input_with_timeout function
            answer = get_input_with_timeout("Enter your answer (A, B, or C): ", 10).upper()

             # Check if the user's answer is correct
            if answer == "A":
                print("\nCongratulations! You solved the riddle and opened the door.")
                break
                
            # Check if the user didn't provide an answer within the timeout    
            elif not answer:
                print("\nTime's up! You didn't provide an answer. Try again.")
            
            # If the answer is incorrect, display an appropriate message and continue the loop
            else:
                print("\nIncorrect answer. The door remains closed. Try again.")
                continue
                
                
        print("\nScenario 2: You enter a room with a locked box.")
        print("To unlock the box, you must solve a riddle.")
        print("\nDavid's father has three sons: Snap, Crackle, and _____?")
        print("\nA) Jhon")
        print("B) Micheal")
        print("C) David")
        
        # Begin a nested loop to repeatedly ask for user input within a timeout
        
        while True:  # Loop for user input with timeout
            
            # Get user input using the get_input_with_timeout function
            answer = get_input_with_timeout("Enter your answer (A, B, or C): ", 10).upper()

            # Check if the user's answer is correct
            if answer == "C":
                print("\nCongratulations! You solved the riddle and unlocked the box.")
                break
            
            # Check if the user didn't provide an answer within the timeout
            elif not answer:
                print("\nTime's up! You didn't provide an answer. Try again.")

            # If the answer is incorrect, display an appropriate message and continue the loop    
            else:
                print("\nIncorrect answer. The box remains locked. Try again.")
                continue

        print("\nScenario 3: You enter a maze with multiple paths.")
        print("To find the exit, you must solve a riddle at each intersection.")
        print("Choose the correct path based on the riddle's answer.")

        print("\nIntersection 1: I am easy to lift, but hard to throw. What am I?")
        print("\nA) Metal")
        print("B) A feather ")
        print("C) Dumble's ")
        
        # Begin a nested loop to repeatedly ask for user input within a timeout
        
        while True:   # Loop for user input with timeout
            
            # Get user input using the get_input_with_timeout function
            answer = get_input_with_timeout("Enter your answer (A, B, or C): ", 10).upper()

            # Check if the user's answer is correct
            if answer == "B":
                print("\nCorrect! Proceed to the next intersection.")
                break
            
            # Check if the user didn't provide an answer within the timeout
            elif not answer:
                print("\nTime's up! You didn't provide an answer. Try again.")

            # If the answer is incorrect, display an appropriate message and continue the loop    
            else:
                print("\nIncorrect answer. You take the wrong path and get lost. Try again.")
                continue
                
        print("\nIntersection 2: I'm tall when I'm young and short when I'm old. What am I?")
        print("\nA) Tree")
        print("B) A candle")
        print("C) Book")
        
        # Begin a nested loop to repeatedly ask for user input within a timeout
        
        while True:   # Loop for user input with timeout
            
            # Get user input using the get_input_with_timeout function
            answer = get_input_with_timeout("Enter your answer (A, B, or C): ", 10).upper()

            # Check if the user's answer is correct
            if answer == "B":
                print("\nCorrect! You found the exit. Well done!")
                break
            
            # Check if the user didn't provide an answer within the timeout
            elif not answer:
                print("\nTime's up! You didn't provide an answer. Try again.")

            # If the answer is incorrect, display an appropriate message and continue the loop    
            else:
                print("\nIncorrect answer. You take the wrong path and get lost. Try again.")
                continue
                
        print("\nScenario 4: You wake up in a room with three doors.")
        print("Each door has a riddle, and only one leads to freedom.")

        print("\nDoor 1: I can run but never walk, I have a mouth but never talk, I have a bed but never sleep. What am I?")
        print("\nA) Road")
        print("B) Wind ")
        print("C) River")
        
        # Begin a nested loop to repeatedly ask for user input within a timeout
        
        while True:    # Loop for user input with timeout
            
            # Get user input using the get_input_with_timeout function
            answer = get_input_with_timeout("Enter your answer (A, B, or C): ", 10).upper()


            time.sleep(10)

            # Check if the user's answer is correct
            if answer == "C":
                print("\nCorrect! You chose the right door and escaped.")
                break
            
            # Check if the user didn't provide an answer within the timeout
            elif not answer:
                print("\nTime's up! You didn't provide an answer. Try again.")
            
            # If the answer is incorrect, display an appropriate message and continue the loop
            else:
                print("\nIncorrect answer. You open the wrong door and remain trapped. Try again.")
                continue

        print("\nScenario 5: You find a mysterious ancient artifact.")
        print("To uncover its secrets, you must solve a riddle.")
        print("\nI'm not alive, but I can die. I have no lungs, but I need air. What am I?")
        print("\nA) Fire")
        print("B) Stone")
        print("C) Plant")
        
        # Begin a nested loop to repeatedly ask for user input within a timeout
        
        while True:    # Loop for user input with timeout
            
             # Get user input using the get_input_with_timeout function
            answer = get_input_with_timeout("Enter your answer (A, B, or C): ", 10).upper()


            time.sleep(10)

            # Check if the user's answer is correct
            if answer == "A":
                print("\nCongratulations! You solved the riddle and unlocked the artifact's secrets.")
                break
            
            # Check if the user didn't provide an answer within the timeout
            elif not answer:
                print("\nTime's up! You didn't provide an answer. Try again.")
            
            # If the answer is incorrect, display an appropriate message and continue the loop
            else:
                    print("\nIncorrect answer. The artifact remains enigmatic. Try again.")
                    continue

        print("\nScenario 6: You encounter a talking statue at a crossroads.")
        print("To gain its guidance, you must answer a riddle.")
        print("\nI'm a word of letters three; add two and fewer there will be.")
        print("\nA) Few")
        print("B) The")
        print("C) Less")
        
        # Begin a nested loop to repeatedly ask for user input within a timeout
        
        while True:   # Loop for user input with timeout
            
             # Get user input using the get_input_with_timeout function
            answer = get_input_with_timeout("Enter your answer (A, B, or C): ", 10).upper()


            time.sleep(10)

            # Check if the user's answer is correct
            if answer == "A":
                print("\nCorrect! The statue imparts its wisdom, guiding you on your journey.")
                break
            
            # Check if the user didn't provide an answer within the timeout
            elif not answer:
                print("\nTime's up! You didn't provide an answer. Try again.")
            
            # If the answer is incorrect, display an appropriate message and continue the loop
            else:
                print("\nIncorrect answer. The statue remains silent. Try again.")
                continue


        # Ask if the player wants to play again        
        play_again = input("\nDo you want to play again? (yes/no): ")
        
        # If the player's answer is not "yes", break out of the loop to end the game
        if play_again.lower() != "yes":
            break

            
    # Display a farewell message when the game ends    
    print("\nThank you for playing the Riddle-solving Game. Goodbye!")

# Check if the script is being run as the main module
if __name__ == "__main__":
    play_game()  # Call the main function to start the game
```
